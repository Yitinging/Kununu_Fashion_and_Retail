```python
import  pandas as pd
import numpy as np
```

```python
df_ku = pd.read_csv('Kununu_Fashion_and_Retail.csv')
```

```python
df_ku.head()
```

```python
# Filter competitor companies
companies = ["Adidas", "Bosch", "Burberry", "PVH", "Ralph Lauren", "Giorgio Armani", "LVMH", "Marc O'Polo","Mercedes-Benz", "Porsche", "Hugo Boss"]

df_ku['company_name_short'] = np.nan

for company in companies:
    df_ku.loc[df_ku['company_name'].str.contains(company, case=False),'company_name_short'] = company
    
    
df_ku = df_ku[df_ku['company_name_short'].isin(companies)]
    
```

```python
df_ku.columns
```

```python
df_ku.shape
```

```python
# drop out rows with null score 
score_columns = ['review_score',
       'employer_atmosphere_score', 'employer_communication_score',
       'employer_teamwork_score', 'employer_work_life_score',
       'employer_leadership_score', 'employer_tasks_score',
       'employer_equality_score', 'employer_old_colleagues_score',
       'employer_work_conditions_score', 'employer_environment_score',
       'employer_salary_score', 'employer_image_score',
       'employer_career_score']

df_ku2 = df_ku.dropna(subset=score_columns)
df_ku2.shape
```
