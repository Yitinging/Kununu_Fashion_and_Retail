```python
# Filter columns that contain the keyword 'text' and also keep 'company_name column'
filtered_columns = [col for col in df.columns if 'text' in col or col == 'company_name']

# Create a new DataFrame with only the filtered columns
df_new = df[filtered_columns]

# Convert all words in the 'company_name' column to lowercase
df_new['company_name'] = df_new['company_name'].str.lower()

df_new.head()
```

```python
# Function to assign values based on company_name
def assign_company_name(row):
    if 'hugo boss' in row:
        return 'Hugo Boss'
    elif 'bosch' in row:
        return 'Bosch'
    elif 'adidas' in row:
        return 'Adidas'
    elif 'giorgio armani' in row:
        return 'Giorgio Armani'
    elif 'lvmh' in row:
        return 'LVMH'
    elif 'marc o' in row:
        return 'Marc O Polo'
    elif 'mercedes' in row:
        return 'Mercedes Benz'
    elif 'porsche' in row:
        return 'Porsche'
    elif 'pvh' in row:
        return 'PVH'
    elif 'ralph lauren' in row:
        return 'Ralph Lauren'
    else:
        return None



# Create a new column based on the company_name values
df_new['company_group'] = df_new['company_name'].apply(assign_company_name)

# Filter out rows where company_group is None
df_filtered = df_new[df_new['company_group'].notnull()]

# Display the filtered DataFrame
df_filtered
```

```python
unique_values = df_filtered['company_group'].unique()
unique_values
```

```python
# Group the data by company_group
grouped_df = df_filtered.groupby('company_group')

# Create a dictionary to store DataFrames for each group
group_dfs = {}

# Iterate over the groups and create a new DataFrame for each group
for name, group in grouped_df:
    group_dfs[name] = group.reset_index(drop=True)
    print(f"\nDataFrame for group: {name}")
    print(group_dfs)
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Function to calculate and display the most frequently used words for each company
def display_most_frequent_words(text_data, company, top_n=10):
    # Split the text into words (tokens)
    words = text_data.split()
    
    # Count word frequencies
    word_frequencies = Counter(words)
    
    # Get the top N most common words
    most_common_words = word_frequencies.most_common(top_n)
    
    # Display the top N most common words
    print(f"Top {top_n} most frequently used words for {company}:")
    for word, freq in most_common_words:
        print(f"{word}: {freq}")
    
    # Plot the top N words in a bar chart
    words, counts = zip(*most_common_words)
    plt.figure(figsize=(7, 3))
    plt.barh(words, counts, color='skyblue')
    plt.title(f"Top {top_n} Most Frequent Words for {company}")
    plt.xlabel('Frequency')
    plt.ylabel('Words')
    plt.show()
# Iterate over each company in the `group_dfs` dictionary
for company, df in group_dfs.items():
    print(f"Processing company: {company}")
    
    # Clean each feedback column and store in new columns
    for col in columns_to_analyze:
        if col in df.columns:
            df[col + '_cleaned'] = df[col].apply(clean_text)

    # Combine all feedback columns into a single column
    df['all_feedback_cleaned'] = df[[
        col + '_cleaned' for col in columns_to_analyze if col + '_cleaned' in df.columns
    ]].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

    # Combine all the cleaned feedback into one string for each company
    combined_feedback = ' '.join(df['all_feedback_cleaned'])
    
    # Display the most frequent words for each company
    display_most_frequent_words(combined_feedback, company, top_n=10)
```
